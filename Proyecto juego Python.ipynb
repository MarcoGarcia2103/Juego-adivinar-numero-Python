```python
#Juego adivina el numero de 1 al 100

#Ficheros necesarios

import getpass
import random
import pandas as pd
import matplotlib.pyplot as plt
import Estadisticas  # Modulo creado por estudiante para guardar resultados del juego (verificar ruta)
import winsound
import matplotlib.image as mpimg

#Funcion sonido
def reproducir_sonido(ruta_archivo):
    winsound.PlaySound(ruta_archivo, winsound.SND_FILENAME)

#Funcion imagenes
def mostrar_imagen(ruta_archivo):
    img = mpimg.imread(ruta_archivo)
    plt.imshow(img)
    plt.axis('off')
    plt.show()

#Nombre del jugador
def nombre_jugador():
    return input('Ingresa nombre del Jugador: \n')

#Menu dificultad
def menu_dificultad():
    print('Elige nivel de juego: \n1. Facil \n2. Medio \n3. Dificil')
    opcion2 = int(input('Elige nivel de dificultad 1 y 3: \n'))
    while opcion2 < 1 or opcion2 > 3:
        opcion2 = int(input('La dificultad debe estar entre 1 y 3: \n'))
    intentos = {1: 20, 2: 12, 3: 5}.get(opcion2, 20)
    print(f'\nNivel {["Facil", "Medio", "Dificil"][opcion2-1]}')
    return opcion2, intentos

#Jugador 1 ingresa numero secreto oculto
def jugador_1():
    print('\nTurno Jugador 1')
    return int(getpass.getpass('Ingresa el número secreto entre 1 y 1000: \n'))

#Opcion dos jugadores
def dos_jugadores(numero_1, intentos_maximos):
    intentos = 0
    while intentos < intentos_maximos:
        valor1 = int(input("Turno jugador 2: Adivina el número entre 1 y 1000: "))
        intentos += 1
        print(f'\nLlevas {intentos} de {intentos_maximos} intentos')
        if valor1 < numero_1:
            print("El número ingresado es Menor \n")
        elif valor1 > numero_1:
            print("El número ingresado es Mayor \n")
        else:
            print("¡Felicidades! Has ganado.")
            reproducir_sonido("C:\\Users\marco\\My Python stuff\\sonido_victoria.wav") #agregar ruta de sonido de victoria
            mostrar_imagen("C:\\Users\marco\\My Python stuff\\imagen_victoria.png") #agregar ruta de imagen de victoria
            return intentos, True
    print("Lo siento, has agotado tus intentos. El número era:", numero_1)
    reproducir_sonido("C:\\Users\marco\\My Python stuff\\sonido_derrota.wav") #agregar ruta de sonido de derrota
    mostrar_imagen("C:\\Users\marco\\My Python stuff\\imagen_derrota.png") #agregar ruta de imagen de derrota
    return intentos, False

#Calcular puntos obtenido segun nivel de dificultad e intentos utilizados
def calcular_puntos(nivel_dificultad, intentos_usados):
    puntos_base = {1: 100, 2: 150, 3: 200}
    penalizacion = 5 * intentos_usados if intentos_usados > 1 else 0
    return max(0, puntos_base.get(nivel_dificultad, 0) - penalizacion)

#Funcion aleatorio para modo solitario
def aleatorio(minimo, maximo):
    return random.randint(minimo, maximo)

#Funcion que compara aleatorio con respuestas
def comparacion_numero(numero_aleatorio, intentos_maximos):
    intentos = 0
    while intentos < intentos_maximos:
        valor1 = int(input("Adivina el número entre 1 y 1000: "))
        intentos += 1
        print(f'Llevas {intentos} de {intentos_maximos} intentos \n')
        if valor1 < numero_aleatorio:
            print("El número ingresado es menor")
        elif valor1 > numero_aleatorio:
            print("El número ingresado es mayor")
        else:
            print("¡Felicidades! Has ganado.")
            reproducir_sonido("C:\\Users\marco\\My Python stuff\\sonido_victoria.wav") #agregar ruta de sonido de victoria
            mostrar_imagen("C:\\Users\marco\\My Python stuff\\imagen_victoria.png") #agregar ruta de imagen de victoria
            return intentos, True
    print("Lo siento, has agotado tus intentos. El número era:", numero_aleatorio)
    reproducir_sonido("C:\\Users\marco\\My Python stuff\\sonido_derrota.wav") #agregar ruta de sonido de derrota
    mostrar_imagen("C:\\Users\marco\\My Python stuff\\imagen_derrota.png") #agregar ruta de imagen de derrota
    return intentos, False

#Menu tipo de partida
def menu_partida():
    print('Elige tipo de partida: \n1. Solitario \n2. Dos jugadores \n3. Ver estadísticas \n4. Salir')
    opcion1 = int(input('Escribe una opción entre 1 y 4: '))
    while opcion1 < 1 or opcion1 > 4:
        opcion1 = int(input('Escribe una opción entre 1 y 4: '))
    return opcion1

#Funcion dos jugadores para guardar en fichero Excel
def comprobacion_jugadores():
    nombre = nombre_jugador()
    nivel_dificultad, intentos_maximos = menu_dificultad()
    numero_1 = jugador_1()
    intentos_usados, exito = dos_jugadores(numero_1, intentos_maximos)
    resultado = "Ganó" if exito else "Perdió"
    puntos = calcular_puntos(nivel_dificultad, intentos_usados)
    print(f"{nombre} obtuviste {puntos} puntos. \n")
    Estadisticas.guardar_resultados(nombre, resultado, intentos_usados, puntos, nivel_dificultad)

#Funcion un jugador para guardar en fichero Excel
def comprobacion_solitario():
    nombre = nombre_jugador()
    nivel_dificultad, intentos_maximos = menu_dificultad()
    numero_aleatorio = aleatorio(1, 1000)
    intentos_usados, exito = comparacion_numero(numero_aleatorio, intentos_maximos)
    resultado = "Ganó" if exito else "Perdió"
    puntos = calcular_puntos(nivel_dificultad, intentos_usados)
    print(f"{nombre} obtuviste {puntos} puntos. \n")
    Estadisticas.guardar_resultados(nombre, resultado, intentos_usados, puntos, nivel_dificultad)

#Funcion para ver estadadisticas del juego
def mostrar_estadisticas():
    archivo_excel = "C:\\Users\\marco\\My Python stuff\\Estadistica.xlsx"
    df = pd.read_excel(archivo_excel)
    
    while True:
        print('\nOpciones de estadísticas: \n1. Top 5 jugadores por puntos \n2. Buscar puntos por nombre \n3. Últimos 5 puestos por puntos \n4. Top 3 jugadores con más victorias \n5. Regresar al menú principal\n')
        opcion = int(input('Elige una opción: \n'))
        
        if opcion == 1:
            top_5 = df.nlargest(5, 'puntos')
            print('\nTop 5 Jugadores por Puntos:\n')
            print(top_5[['nombre', 'puntos']], '\n')
            top_5[['nombre', 'puntos']].set_index('nombre').plot(kind='bar', title='Top 5 Jugadores por Puntos', legend=False)
            plt.xlabel('Nombre')
            plt.ylabel('Puntos')
            plt.show()
        
        elif opcion == 2:
            nombre = input('\nIngresa el nombre del jugador: \n')
            jugador = df[df['nombre'] == nombre]
            if not jugador.empty:
                print(f'\nInformación del Jugador {nombre}:\n')
                print(jugador[['nombre', 'puntos']], '\n')
            else:
                print(f'\nNo se encontraron registros para el jugador: {nombre}\n')
        
        elif opcion == 3:
            bottom_5 = df.nsmallest(5, 'puntos')
            print('\nÚltimos 5 Jugadores por Puntos:\n')
            print(bottom_5[['nombre', 'puntos']], '\n')
            bottom_5[['nombre', 'puntos']].set_index('nombre').plot(kind='bar', title='Últimos 5 Jugadores por Puntos', color='orange', legend=False)
            plt.xlabel('Nombre')
            plt.ylabel('Puntos')
            plt.show()
        
        elif opcion == 4:
            top_3_victorias = df[df['resultado'] == 'Ganó']['nombre'].value_counts().head(3)
            print('\nTop 3 Jugadores con Más Victorias:\n')
            print(top_3_victorias, '\n')
            top_3_victorias.plot(kind='bar', title='Top 3 Jugadores con Más Victorias', color='green', legend=False)
            plt.xlabel('Nombre')
            plt.ylabel('Número de Victorias')
            plt.show()
        
        elif opcion == 5:
            print('\nRegresando al menú principal...\n')
            return
        
        else:
            print('\nOpción no válida. Intenta de nuevo.\n')

#Funcion principal del juego
def main():
    while True:
        opcion = menu_partida()
        if opcion == 1:
            comprobacion_solitario()
        elif opcion == 2:
            comprobacion_jugadores()
        elif opcion == 3:
            mostrar_estadisticas()
        elif opcion == 4:
            print("Gracias por jugar, vuelve pronto")
            return

if __name__ == "__main__":
    main()
```
